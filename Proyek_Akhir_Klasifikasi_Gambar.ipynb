```python
import zipfile, os        
local_zip='/content/rockpaperscissors.zip'  
zip_ref= zipfile.ZipFile(local_zip, 'r')  
zip_ref.extractall('proyek_akhir')       
zip_ref.close()             
```

```python
folder_rock = os.path.join('/content/proyek_akhir/rockpaperscissors/rock')   
folder_paper = os.path.join('/content/proyek_akhir/rockpaperscissors/paper') 
folder_scissors = os.path.join('/content/proyek_akhir/rockpaperscissors/scissors')


print('Jumlah gambar Rock:', len(os.listdir(folder_rock)))    
print('Jumlah gambar Paper:', len(os.listdir(folder_paper)))    
print('Jumlah gambar Scissors:', len(os.listdir(folder_scissors)))


filename_rock = os.listdir(folder_rock) 
print("\nNama file Rock:", filename_rock[:3])              
filename_paper = os.listdir(folder_paper)
print("Nama file Paper:", filename_paper[:3]) 
filename_scissors = os.listdir(folder_scissors)
print("Nama file Scissors:", filename_scissors[:3])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


nbaris = 3                                                                  
nkolom = 3                                                                    

picture_index = 0                                                            

fig = plt.gcf()
fig.set_size_inches(nkolom * 3, nbaris * 3)                                   

picture_index += 3                                                                
gambar_paper  = [os.path.join(folder_paper, fname)                       
                      for fname in filename_paper[picture_index-3:picture_index]]
gambar_rock = [os.path.join(folder_rock, fname)                        
                      for fname in filename_rock[picture_index-3:picture_index]]
gambar_scissors = [os.path.join(folder_scissors, fname)                       
                      for fname in filename_scissors[picture_index-3:picture_index]]

for i, img_path in enumerate(gambar_paper+gambar_rock+gambar_scissors):
  sp = plt.subplot(nbaris, nkolom, i + 1)                                        
  sp.axis('Off')                                                                

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```

```python
import tensorflow as tf 
import keras_preprocessing 
from keras_preprocessing import image
from keras_preprocessing.image import ImageDataGenerator
from tensorflow.python.ops.batch_ops import batch


FOLDER_UTAMA='/content/proyek_akhir/rockpaperscissors/rps-cv-images'
training_datagen = ImageDataGenerator( 
      rescale=1./255,
      rotation_range=40, 
      width_shift_range=0.2, 
      height_shift_range=0.2,       
      shear_range=0.2,     
      zoom_range=0.2,  
      horizontal_flip=True,
      vertical_flip = True,
      fill_mode = 'nearest', 
      validation_split=0.4)

training_generator = training_datagen.flow_from_directory(
    FOLDER_UTAMA,
    target_size=(150,150),
    color_mode='rgb',
    class_mode='categorical',
    shuffle=True,
    subset='training')

validation_generator = training_datagen.flow_from_directory(
    FOLDER_UTAMA,
    target_size=(150,150),
    color_mode='rgb',
    class_mode='categorical',
    shuffle=True,
    subset='validation')
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2), 

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),          

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),

    tf.keras.layers.Dense(3, activation='softmax')
])

model.summary()

model.compile(loss='categorical_crossentropy',                                  
              optimizer='adam',
              metrics=['accuracy'])
```

```python
histori= model.fit(
    training_generator,
    steps_per_epoch=4,
    epochs=25,
    validation_data= validation_generator,
    validation_steps=4,
    verbose=2)
```

```python
import matplotlib.pyplot as plt

acc=histori.history['accuracy']
val_acc= histori.history['val_accuracy']
loss = histori.history['loss']
val_loss= histori.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Akurasi Training')
plt.plot(epochs, val_acc, 'g', label='Akurasi Validasi')
plt.title('Akurasi hasil training dan validasi')
plt.legend(loc=0)
plt.figure()
plt.show()

plt.plot(epochs, loss, 'r', label='Loss Training')
plt.plot(epochs, val_loss, 'g', label='Loss Validasi')
plt.title('Loss hasil training dan validasi')
plt.legend(loc=0)
plt.figure()
plt.show()
```

# Uji Coba Model

```python
import numpy as np
from google.colab import files
from tensorflow.keras.preprocessing import image

upload_gambar = files.upload()
for fn in upload_gambar.keys():
  #memprediksi gambar
  path = fn
  gambar= image.load_img(path, target_size=(150,150))

  plot_gambar=plt.imshow(gambar)
  x=image.img_to_array(gambar)
  x=np.expand_dims(x, axis=0)
  images=np.vstack([x])

  classes=model.predict(images, batch_size=10)
  print(fn)
  if classes[0][0]==1:
    print('Tangan berbentuk "Paper"')
  elif classes[0][1]==1:
    print('Tangan berbentuk "Rock"')
  elif classes[0][2]==1:
    print('Tangan berbentuk "Scissors"')
  else:
    print('Bentuk tidak diketahui')
```

# Klasifikasi Gambar dengan Kriteria Callback

```python
class newCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('val_accuracy')>0.9):
      print("\nPelatihan model berhenti. \n Akurasi model telah melebihi 90%. ")
      self.model.stop_training=True
callbacks=newCallback()

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2), 

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),   

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),

    tf.keras.layers.Dense(3, activation='softmax')
])

model.summary()

model.compile(loss='categorical_crossentropy',                                  
              optimizer='rmsprop',
              metrics=['accuracy'])

histori= model.fit(
    training_generator,
    steps_per_epoch=20,
    epochs=100,
    validation_data= validation_generator,
    validation_steps=4,
    verbose=2,
    callbacks=[callbacks])

import matplotlib.pyplot as plt

acc=histori.history['accuracy']
val_acc= histori.history['val_accuracy']
loss = histori.history['loss']
val_loss= histori.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Akurasi Training')
plt.plot(epochs, val_acc, 'y', label='Akurasi Validasi')
plt.title('Hasil Akurasi training dan validasi')
plt.legend(loc=0)
plt.figure()
plt.show()

plt.plot(epochs, loss, 'b', label='Loss Training')
plt.plot(epochs, val_loss, 'y', label='Loss Validasi')
plt.title('Hasil loss training dan validasi')
plt.legend(loc=0)
plt.figure()
plt.show()
```
